```python
import math
import numpy as np
import matplotlib.pyplot as plt

A = 5
k = 1
m = 1
f = 1
omega = math.sqrt(k/m)
```

```python
dt = 0.001
t = np.arange(0, 100, dt)
x_t = []
vx_t = []
ax_t = []
for item in t:
    x_t.append(A*math.cos(omega*item))
    vx_t.append(-omega*A*math.sin(omega*item))
    ax_t.append(-omega*omega*A*math.cos(omega*item))
EK_t = []
EP_t = []
EC_t = []
for i in range(len(t)):
    EK_t.append(0.5*m*vx_t[i]**2)
    EP_t.append(0.5*k*(x_t[i])**2)
    EC_t.append(EK_t[i]+EP_t[i])
```

```python
plt.plot(t, x_t, '-b', t, vx_t, '-r', t, ax_t, '-g')
plt.xlim(0, 25)
plt.ylim(-10, 10)
plt.xlabel('t', fontsize = '14')
plt.ylabel('x(t), v(t), a(t)', fontsize = '14')
plt.legend(['x(t)', 'v(t)', 'a(t)'])
plt.grid(color = 'grey', linestyle = '--'
, linewidth = 0.2)
plt.show()
```

```python
plt.plot(t, EC_t, '-b', t, EP_t, '-r', t, EK_t, '-g')
plt.xlim(0, 25)
#plt.ylim(-10, 10)
plt.xlabel('t', fontsize = '14')
plt.ylabel('EC, EP, EK', fontsize = '14')
plt.legend(['EC', 'EP', 'EK'])
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.2)
plt.show()
```

```python
dt = 0.01
t = np.arange(0, 100, dt)
x_s = []
x_s.append(A)
x_s.append(math.sqrt(A**2 - omega*omega*(A*dt)**2))
vx_s = []
vx_s.append(0)
ax_s = []
ax_s.append(-omega*omega*x_s[0])
ax_s.append(-omega*omega*x_s[1])
for i in range(2, len(t)):
    x_s.append(2*x_s[i-1]-x_s[i-2]+(dt**2)*ax_s[i-1])
    ax_s.append(-omega*omega*x_s[i])
    vx_s.append((x_s[i]-x_s[i-2])/(2*dt))
ostatni_x = 2*x_s[i-1]-x_s[i-2]+(dt**2)*ax_s[i-1]
vx_s.append((ostatni_x-x_s[i-2])/(2*dt))
EK_s = []
EP_s = []
EC_s = []
for i in range(len(x_s)):
    EK_s.append(0.5*m*vx_s[i]**2)
    EP_s.append(0.5*k*(x_s[i])**2)
    EC_s.append(EK_s[i]+EP_s[i])
```

```python
plt.plot(t, x_s, '-b', t, vx_s, '-r', t, ax_s, '-g')
plt.xlim(0, 25)
plt.ylim(-10, 10)
plt.xlabel('t', fontsize = '14')
plt.ylabel('x(t), v(t), a(t)', fontsize = '14')
plt.legend(['x(t)', 'v(t)', 'a(t)'])
plt.grid(color = 'grey', linestyle = '--', linewidth = 0.2)
plt.show()
```
